```python
#Carga del clasificador de cascada
import cv2
from google.colab.patches import cv2_imshow
faceClassifier = cv2.CascadeClassifier('/content/haarcascade_frontalcatface.xml')

image = cv2.imread('/Gato.jpg')

#se quita el color y griss
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = faceClassifier.detectMultiScale(gray,
                                     scaleFactor=1.1,
                                     minNeighbors=5,
                                     minSize=(30, 30),
                                     maxSize=(200, 200)
                                  )
print(faces)

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

cv2_imshow(image)
cv2.waitKey()
cv2.destroyAllWindows()

```
