```python
image_path = '/Gato.jpg'
imagen = cv2.imread(image_path)

# define el kernel de convolucion
kernel = np.array(
    [[0, -1, 0],
     [-1, 5, -1],
     [0, -1, 0]])

#aplicar la convolucion

convolved_image = cv2.filter2D(imagen, -1, kernel)

#Mostra l a imagen original y la convolucionada
cv2_imshow(imagen)
cv2_imshow(convolved_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Transformaciones Geometricas

```python
import cv2
import numpy as np

#Cargar la imagen
image_path = '/Gato.jpg'
imagen = cv2.imread(image_path)

#Obtener dimenciones de la imagen
height, width = imagen.shape[:2]

#Obtener el centro de la imagen
center = (width / 2, height / 2)

#Definir la matriz de rotacion
angle = 45
M = cv2.getRotationMatrix2D(center,angle,1.0)

rotated = cv2.warpAffine(imagen,M,(width,height))

#Mostrar la imagen original y la rotada
cv2_imshow(rotated)

cv2.waitKey(0)
cv2.destroyAllWindows()


```

Filtrado de Imagenes

```python
import cv2
#Cargar la imagen
image_path = '/Gato.jpg'
imagen = cv2.imread(image_path)

#Aplicar filtro Gaussiano suavizacion imagen
smoothed = cv2.GaussianBlur(imagen, (5, 5), 0)

#Mostrar la imagen original y la suavizada
cv2_imshow(smoothed)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Filtrado de imagen

```python
#Cargar la imagen
image_path = '/Gato.jpg'
imagen = cv2.imread(image_path)

kernel = np.array(
    [
        [-1, -1, -1],
        [-1,  9, -1],
        [-1, -1, -1]
    ]
)

#Aplicar filtro de afilado
sharpened = cv2.filter2D(imagen, -1, kernel)

#Mostrar la imagen original y la suavizada
cv2_imshow(sharpened)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Deteccion de bordes

```python
#Cargar la imagen
image_path = '/Gato.jpg'
imagen = cv2.imread(image_path)

#Aplicar operador sobel para deteccion de bordes
sobelx = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(imagen, cv2.CV_64F, 0, 1, ksize=5)

#Combinar las respuestas en magnitud
edges = cv2.magnitude(sobelx, sobely)

#Normalizar los valores para mostrar la imagen correctamente
edges = cv2.normalize(edges, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)



#Mostrar la imagen original y la suavizada
cv2_imshow(edges)

cv2.waitKey(0)
cv2.destroyAllWindows()

```
