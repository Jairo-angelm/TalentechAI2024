# Introducción keras en NLP

Ejemplo 1

```python
import keras
import tensorflow as tf
frases= ['Hola mundo',
	'Hola a todos',
	'Hola a todo el mundo'
]
```

Tokenización con keras

Generar el diccionario de tokens

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10)
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print(word_index)
```

Tokenizer para convertir las frases en secuencias de tokens.

```python
secuencias=tokenizer.texts_to_sequences(frases)
print('secuencias = ',secuencias)
```

Ejemplo de Tokenización con keras
método "pad_sequences" de la clase preprocessing.sequence para rellenar las secuencias de tokens a una longitud uniforme.

```python
relleno = tf.keras.preprocessing.sequence.pad_sequences(secuencias)
print('relleno =\n ',relleno)

```

Ejemplo 2

```python
import keras
import tensorflow as tf
frases= ['Hola mundo',
	'Hola a todos',
	'Hola a todo el mundo',
	'Feliz inicio de jornada'
]



tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10)
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print(word_index)


secuencias=tokenizer.texts_to_sequences(frases)
print('secuencias = ',secuencias)

relleno = tf.keras.preprocessing.sequence.pad_sequences(secuencias)
print('relleno =\n ',relleno)

```

Ejemplo 3
El parámetro oov_token (out of vocabulary token)
En el Tokenizer de Keras se utiliza para especicar un token que se asignará a las palabras fuera del vocabulario durante el proceso de tokenización.

```python
import keras
import tensorflow as tf
frases= ['Hola mundo',
	'Hola a todos',
	'Hola a todo el mundo'
	'Feliz inicio de jornada para todos'
]


tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10,oov_token="<OOV>")
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print(word_index)


secuencias=tokenizer.texts_to_sequences(frases)
print('secuencias = ',secuencias)

relleno = tf.keras.preprocessing.sequence.pad_sequences(secuencias)
print('relleno =\n ',relleno)
```

Ejemplo 4


```python
import keras
import tensorflow as tf
frases= ['Hola mundo',
	'Hola a todos',
	'Hola a todo el mundo'
	'Buen dia, Feliz inicio de jornada para todos'
]


tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10,oov_token="<OOV>")
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print(word_index)


secuencias=tokenizer.texts_to_sequences(frases)
print('secuencias = ',secuencias)


relleno = tf.keras.preprocessing.sequence.pad_sequences(secuencias,
                                                        padding='post',
                                                        truncating='post')

print('relleno =\n ',relleno)
```
