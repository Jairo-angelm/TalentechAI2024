# Pronosticos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

Se genera datos de ejemplo

```python
#genera datos de ejemplo
np.random.seed(0)
dates = pd.date_range(start='2022-01-01', end='2022-12-31')
values = np.arange(len(dates))*0.2 + np.random.randn(len(dates))*50
ts = pd.Series(values, index=dates)
```

Dividir los datos en entrenamiento y prueba

```python
train_data = ts[:'2022-10-01']
test_data = ts['2022-10-02':]
```

Ajustar el modelo

```python
model = ARIMA(train_data, order=(10,1,10))
model_fit = model.fit()
```

Realizar pronostico

```python
forecast = model_fit.forecast(steps=len(test_data))
```

Graficas

```python
plt.figure(figsize=(10,6))
plt.plot(train_data, label='Datos Entrenamiento')
plt.plot(test_data, label='Datos Prueba', color ='green')
plt.plot(test_data.index, forecast, label='Pronostico', color='red')
plt.legend()
plt.title("Pronostico de series temporales")
plt.xlabel("Fecha")
plt.show()
```
