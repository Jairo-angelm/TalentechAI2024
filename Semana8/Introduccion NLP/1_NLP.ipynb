Ejemplo básico de clasificacion de texto utilizando el clasificador Naive Bayes de NLTK

1. Importacion de bibliotecas

```python
import nltk
import random
```

Defincion delconjunto de datos etiquetados

```python
data = [("I love this movie"  , "positive"),
    ("This movie is terrible"  , "negative"),
    ("This movie is great"  , "positive"),
    ("I feel amazing today", "positive"),
    ("This movie is great", "positive"),
    ("This film is amazing", "positive"),
    ("I can't stand watching this movie", "negative"),
    ("The acting in this movie is phenomenal", "positive"),
    ("This movie is awful", "negative"),
    ("I don't like this movie", "negative")
    ("this movie lacks depth and sustance", "negative")
    ("The plot of this movie was captivating", "positive")
    ("I found the characters in this film to be very engaging", "positive")
    ("the dialog felt forced and unnatural", "negative")
    ("this movie execeeded my expectation", "positive")
    ("the performance by the actors were lackluster", "negative")
    ]

```

 Preprocesamiento de datos

```python
def preprocess(text):
  tokens = word_tokenize(text.lower())
  return {word: True for word in tokens}
```

Aplicación del preprocesamiento a los datos

```python
featuresets = [(preprocess(text), label) for (text, label) in data]
```

División de datos

```python
train_set, test_set = featuresets[:16], featuresets[:5]
```

Entrenamiento del clasificador

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

Evaluación del clasi

```python
accuracy = nltk.classify.accuracy(classifier, test_set)
print("Accuracy:", accuracy)
```

Clasificacion de un nuevo texto

```python
new_text = "This movie is amaizing"
new_text_features = preprocess(new_text)
prediction = classifier.classify(new_text_features)
print("Prediction:", prediction)
```
